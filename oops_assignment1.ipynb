1) Bank Account Create a class representing a bank account with attributes like account number, account holder name, and balance. Implement methods to deposit and withdraw money from the account.



```python
# Define a class called BankAccount to represent a bank account.
class BankAccount:
    # Constructor method (__init__) initializes the account with account number,
    # account holder name, and an optional balance (defaulted to 0 INR).
    def __init__(self, account_number, account_holder_name, balance=0):
        self.account_number = account_number
        self.account_holder_name = account_holder_name
        self.balance = balance

    # Method to deposit money into the account.
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount  # Increase the account balance.
            print(f"Deposited {amount} INR. New balance: {self.balance} INR")
        else:
            print("Invalid deposit amount. Amount should be greater than 0 INR.")

    # Method to withdraw money from the account (if sufficient balance).
    def withdraw(self, amount):
        if amount > 0 and amount <= self.balance:
            self.balance -= amount  # Decrease the account balance.
            print(f"Withdrew {amount} INR. New balance: {self.balance} INR")
        elif amount <= 0:
            print("Invalid withdrawal amount. Amount should be greater than 0 INR.")
        else:
            print("Insufficient balance for withdrawal.")

    # Method to display the account holder name, account number, and current balance.
    def display_balance(self):
        print(f"Account Holder: {self.account_holder_name}")
        print(f"Account Number: {self.account_number}")
        print(f"Balance: {self.balance} INR")

# Create an instance (object) of the BankAccount class with account details.
account1 = BankAccount("1234567890", "John Doe")

# Deposit and withdraw money from the account.
account1.deposit(5000)
account1.withdraw(2000)

# Display the account's information including the balance.
account1.display_balance()
```

2.Employee Management Create a class representing an employee with attributes like employee ID, name, and salary. Implement methods to calculate the yearly bonus and display employee details.

```python
# Define a class called Employee to represent an employee.
class Employee:
    # Constructor method (__init__) initializes the employee with employee ID, name, and salary.
    def __init__(self, employee_id, name, salary):
        self.employee_id = employee_id
        self.name = name
        self.salary = salary

    # Method to calculate the yearly bonus based on a provided bonus percentage.
    def calculate_yearly_bonus(self, bonus_percentage):
        if bonus_percentage >= 0:
            # Calculate the yearly bonus as a percentage of the salary.
            yearly_bonus = (bonus_percentage / 100) * self.salary
            return yearly_bonus
        else:
            # If the bonus percentage is negative, return 0 bonus.
            return 0

    # Method to display the employee's details, including employee ID, name, and salary.
    def display_employee_details(self):
        print(f"Employee ID: {self.employee_id}")
        print(f"Name: {self.name}")
        print(f"Salary: {self.salary} INR")

# Create an instance (object) of the Employee class with employee details.
employee1 = Employee(1, "Atul", 60000)

# Calculate the yearly bonus with a 10% bonus percentage.
bonus_percentage = 10  # 10% bonus
yearly_bonus = employee1.calculate_yearly_bonus(bonus_percentage)
print(f"Yearly Bonus: {yearly_bonus} INR")

# Display the employee's details, including ID, name, and salary.
employee1.display_employee_details()
```

3.Vehicle Rental Create a class representing a vehicle rental system. Implement methods to rent a vehicle, return a vehicle, and display available vehicles.

```python
class VehicleRentalSystem:
    def __init__(self):
        self.available_vehicles = {}

    def rent_vehicle(self, vehicle_type, vehicle_id):
        if vehicle_type in self.available_vehicles and vehicle_id in self.available_vehicles[vehicle_type]:
            self.available_vehicles[vehicle_type].remove(vehicle_id)
            print(f"Rented {vehicle_type} with ID {vehicle_id}")
        else:
            print(f"{vehicle_type} with ID {vehicle_id} is not available for rent.")

    def return_vehicle(self, vehicle_type, vehicle_id):
        if vehicle_type in self.available_vehicles:
            self.available_vehicles[vehicle_type].append(vehicle_id)
            print(f"Returned {vehicle_type} with ID {vehicle_id}")
        else:
            self.available_vehicles[vehicle_type] = [vehicle_id]
            print(f"Returned {vehicle_type} with ID {vehicle_id}")

    def display_available_vehicles(self):
        for vehicle_type, vehicles in self.available_vehicles.items():
            print(f"Available {vehicle_type}s: {', '.join(vehicles)}")

# Create an instance of the VehicleRentalSystem
rental_system = VehicleRentalSystem()

# Add available vehicles
rental_system.available_vehicles['Car'] = ['CAR001', 'CAR002', 'CAR003']
rental_system.available_vehicles['Bike'] = ['BIKE001', 'BIKE002']

# Rent and return vehicles
rental_system.rent_vehicle('Car', 'CAR001')
rental_system.rent_vehicle('Bike', 'BIKE003')  # This bike is not available.
rental_system.return_vehicle('Car', 'CAR001')
rental_system.return_vehicle('Bike', 'BIKE003')  # This bike will be added as available.

# Display available vehicles
rental_system.display_available_vehicles()
```

4.Library Catalog Create classes representing a library and a book. Implement methods to add books to the library, borrow books, and display available books.

```python
# Define a class called Book to represent individual books in the library.
class Book:
    def __init__(self, book_id, title, author):
        self.book_id = book_id  # Unique identifier for the book.
        self.title = title
        self.author = author
        self.is_available = True  # A new book is initially available.

    # Method to borrow a book.
    def borrow(self):
        if self.is_available:
            self.is_available = False  # Set the book as not available.
            return True
        else:
            return False  # Return False if the book is already borrowed.

    # Method to return a book.
    def return_book(self):
        if not self.is_available:
            self.is_available = True  # Set the book as available.
            return True
        else:
            return False  # Return False if the book is already available.

    # Custom string representation for the book.
    def __str__(self):
        availability = "Available" if self.is_available else "Not Available"
        return f"Book ID: {self.book_id}, Title: {self.title}, Author: {self.author}, Status: {availability}"

# Define a class called Library to manage a collection of books.
class Library:
    def __init__(self):
        self.books = {}  # A dictionary to store books with book_id as keys.

    # Method to add a book to the library.
    def add_book(self, book):
        self.books[book.book_id] = book
        print(f"Added Book - {book.title}")

    # Method to borrow a book from the library.
    def borrow_book(self, book_id):
        if book_id in self.books:
            book = self.books[book_id]
            if book.borrow():
                print(f"Borrowed Book - {book.title}")
            else:
                print(f"Book - {book.title} is already borrowed.")
        else:
            print("Book not found in the library.")

    # Method to return a borrowed book to the library.
    def return_book(self, book_id):
        if book_id in self.books:
            book = self.books[book_id]
            if book.return_book():
                print(f"Returned Book - {book.title}")
            else:
                print(f"Book - {book.title} is already available in the library.")
        else:
            print("Book not found in the library.")

    # Method to display available books in the library.
    def display_available_books(self):
        print("Available Books in the Library:")
        for book_id, book in self.books.items():
            if book.is_available:
                print(book)

# Create an instance of the Library class
library = Library()

# Add books to the library
book1 = Book(1, "The Great Gatsby", "F. Scott Fitzgerald")
book2 = Book(2, "To Kill a Mockingbird", "Harper Lee")
library.add_book(book1)
library.add_book(book2)

# Borrow and return books
library.borrow_book(1)
library.borrow_book(1)  # This book is already borrowed.
library.return_book(1)
library.return_book(2)  # This book is already available.

# Display available books in the library
library.display_available_books()
```

5.Product Inventory Create classes representing a product and an inventory system. Implement methods to add products to the inventory, update product quantity, and display available products.

```python
# Define a class called Product to represent individual products.
class Product:
    def __init__(self, product_id, name, price, quantity):
        self.product_id = product_id  # Unique identifier for the product.
        self.name = name  # Name of the product.
        self.price = price  # Price of the product in USD.
        self.quantity = quantity  # Quantity of the product in the inventory.

    # Method to update the quantity of the product.
    def update_quantity(self, new_quantity):
        self.quantity = new_quantity

    # Custom string representation for the product.
    def __str__(self):
        return f"Product ID: {self.product_id}, Name: {self.name}, Price: {self.price} USD, Quantity: {self.quantity}"

# Define a class called Inventory to manage a collection of products.
class Inventory:
    def __init__(self):
        self.products = []  # A list to store products.

    # Method to add a product to the inventory.
    def add_product(self, product):
        self.products.append(product)

    # Method to update the quantity of a product based on its ID.
    def update_product_quantity(self, product_id, new_quantity):
        for product in self.products:
            if product.product_id == product_id:
                product.update_quantity(new_quantity)

    # Method to display available products in the inventory.
    def display_available_products(self):
        print("Available Products in the Inventory:")
        for product in self.products:
            print(product)

# Create an instance of the Inventory class.
inventory = Inventory()

# Add products to the inventory.
product1 = Product(1, "Laptop", 1000, 10)
product2 = Product(2, "Smartphone", 500, 20)
inventory.add_product(product1)
inventory.add_product(product2)

# Update the quantity of a product.
inventory.update_product_quantity(1, 8)

# Display available products in the inventory.
inventory.display_available_products()
```

6.Shape Calculation Create a class representing a shape with attributes like length, width, and height. Implement methods to calculate the area and perimeter of the shape.

```python
class Shape:
    def __init__(self, length, width=None, height=None):
        self.length = length
        self.width = width
        self.height = height

    def calculate_area(self):
        raise NotImplementedError("Subclasses must implement this method")

    def calculate_perimeter(self):
        raise NotImplementedError("Subclasses must implement this method")

class Rectangle(Shape):
    def calculate_area(self):
        if self.width is None:
            return self.length * self.length  # For squares
        else:
            return self.length * self.width

    def calculate_perimeter(self):
        if self.width is None:
            return 4 * self.length  # For squares
        else:
            return 2 * (self.length + self.width)

class Circle(Shape):
    def calculate_area(self):
        return 3.14159265359 * self.length**2

    def calculate_perimeter(self):
        return 2 * 3.14159265359 * self.length

class Triangle(Shape):
    def calculate_area(self):
        if self.width is None or self.height is None:
            raise ValueError("Both width and height are required for triangle area calculation")
        return 0.5 * self.length * self.width

    def calculate_perimeter(self):
        if self.width is None or self.height is None:
            raise ValueError("Both width and height are required for triangle perimeter calculation")
        return self.length + self.width + self.height

# Example usage:
rectangle = Rectangle(5, 4)
circle = Circle(3)
triangle = Triangle(5, 4, 3)

print("Rectangle - Area:", rectangle.calculate_area(), "Perimeter:", rectangle.calculate_perimeter())
print("Circle - Area:", circle.calculate_area(), "Circumference:", circle.calculate_perimeter())
print("Triangle - Area:", triangle.calculate_area(), "Perimeter:", triangle.calculate_perimeter())
```

7.Student Management Create a class representing a student with attributes like student ID, name, and grades. Implement methods to calculate the average grade and display student details.

```python
# Define a class named Student
class Student:
    # Constructor to initialize the student object with student_id and name
    def __init__(self, student_id, name):
        self.student_id = student_id  # Initialize student ID attribute
        self.name = name  # Initialize name attribute
        self.grades = []  # Initialize an empty list to store student's grades

    # Method to add a grade to the student's list of grades
    def add_grade(self, grade):
        self.grades.append(grade)

    # Method to calculate the average grade of the student
    def calculate_average_grade(self):
        if len(self.grades) == 0:
            return 0  # Return 0 if there are no grades to calculate an average
        total = sum(self.grades)  # Calculate the sum of grades
        return total / len(self.grades)  # Calculate and return the average grade

    # Method to display student details, including student ID, name, grades, and average grade
    def display_student_details(self):
        print(f"Student ID: {self.student_id}")
        print(f"Name: {self.name}")
        print("Grades:", ", ".join(map(str, self.grades)))# Convert grades to a string for display
        print(f"Average Grade: {self.calculate_average_grade()}")

# Example usage:
if __name__ == "__main__":
    # Create two student objects
    student1 = Student("12345", "Raj Mathur")
    student1.add_grade(85)
    student1.add_grade(92)
    student1.add_grade(78)

    student2 = Student("54321", "Utsav Jain")
    student2.add_grade(90)
    student2.add_grade(88)
    student2.add_grade(94)

    # Display the details of the first student
    student1.display_student_details()
    print()  # Add a blank line for separation
    # Display the details of the second student
    student2.display_student_details()
```

8. Email Management Create a class representing an email with attributes like sender, recipient, and subject. Implement methods to send an email and display email details.

```python
# Define a class named Email
class Email:
    # Constructor to initialize the email object with sender, recipient, and subject
    def __init__(self, sender, recipient, subject):
        self.sender = sender  # Initialize sender attribute
        self.recipient = recipient  # Initialize recipient attribute
        self.subject = subject  # Initialize subject attribute
        self.message = None  # Initialize message attribute as None

    # Method to send an email by setting the message and displaying email information
    def send_email(self, message):
        self.message = message  # Set the message for the email
        # Display email information, including sender, recipient, subject, and the message
        print(f"Email sent from {self.sender} to {self.recipient}")
        print(f"Subject: {self.subject}")
        print("Message:")
        print(message)

    # Method to display email details, including sender, recipient, subject, and the message (if set)
    def display_email_details(self):
        print(f"Sender: {self.sender}")
        print(f"Recipient: {self.recipient}")
        print(f"Subject: {self.subject}")
        if self.message:
            print("Message:")
            print(self.message)
        else:
            print("No message has been set.")

# Create an Email object
email = Email("pwskills@example.com", "Rohan@example.com", "Hello, Rohan!")

# Send an email with a message
email.send_email("This is a test email from pwskills to Rohan.")

# Display email details
email.display_email_details()
```

9.Social Media Profile Create a class representing a social media profile with attributes like username and posts. Implement methods to add posts, display posts, and search for posts by keyword

```python
class SocialMediaProfile:
    def __init__(self, username):
        self.username = username
        self.posts = []

    def add_post(self, post):
        self.posts.append(post)

    def display_posts(self):
        if not self.posts:
            print(f"No posts found for {self.username}.")
        else:
            print(f"Posts for {self.username}:")
            for post in self.posts:
                print(post)

    def search_posts_by_keyword(self, keyword):
        matching_posts = [post for post in self.posts if keyword in post]
        if not matching_posts:
            print(f"No posts containing '{keyword}' found for {self.username}.")
        else:
            print(f"Posts containing '{keyword}' for {self.username}:")
            for post in matching_posts:
                print(post)

# Create a social media profile for a user
profile = SocialMediaProfile("Ayush")

# Add some posts to the profile
profile.add_post("Just had a great day at the park!")
profile.add_post("Trying out a new recipe for dinner tonight.")
profile.add_post("Enjoying the latest book I'm reading.")

# Display all the posts in the profile
profile.display_posts()

# Search for posts containing a keyword
profile.search_posts_by_keyword("great")
```

10.ToDo List Create a class representing a ToDo list with attributes like tasks and due dates. Implement methods to add tasks, mark tasks as completed, and display pending tasks.

```python
class ToDoList:
    def __init__(self):
        self.tasks = {}  # A dictionary to store tasks and their due dates

    def add_task(self, task, due_date):
        self.tasks[task] = due_date

    def mark_completed(self, task):
        if task in self.tasks:
            del self.tasks[task]

    def display_pending_tasks(self):
        if not self.tasks:
            print("No pending tasks.")
        else:
            print("Pending Tasks:")
            for task, due_date in self.tasks.items():
                print(f"Task: {task}, Due Date: {due_date}")


to_do_list = ToDoList()

# Add tasks to the to-do list
to_do_list.add_task("Buy groceries", "2023-10-15")
to_do_list.add_task("Complete homework", "2023-10-20")
to_do_list.add_task("Call a friend", "2023-10-25")

# Display pending tasks
to_do_list.display_pending_tasks()

# Mark a task as completed
to_do_list.mark_completed("Buy groceries")

# Display pending tasks after marking one as completed
to_do_list.display_pending_tasks()
```
